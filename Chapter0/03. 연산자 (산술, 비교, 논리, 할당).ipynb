## 3. 연산자 (산술, 비교, 논리, 할당)

#### 기본 설명

연산자는 **변수와 값**에 대한 **연산**을 수행하는 기호입니다. 파이썬은 다양한 종류의 연산자를 제공합니다.

#### 산술 연산자

수학적 계산을 수행합니다.

```python
# 기본 산술 연산자
a = 10
b = 3

print(a + b)  # 덧셈: 13
print(a - b)  # 뺄셈: 7
print(a * b)  # 곱셈: 30
print(a / b)  # 나눗셈: 3.3333333333333335 (항상 float 반환)
print(a // b)  # 정수 나눗셈(몫): 3
print(a % b)  # 나머지: 1
print(a ** b)  # 거듭제곱: 1000 (10의 3승)
```

#### 복합 연산 예제

```python
# 복합 수식
result = (a + b) * 2 - 5
print(result)  # (13 * 2) - 5 = 21

# 우선순위: 괄호 > 지수 > 곱셈/나눗셈 > 덧셈/뺄셈
calculation = 2 + 3 * 4 ** 2
print(calculation)  # 2 + 3 * 16 = 2 + 48 = 50
```

#### 비교 연산자

두 값을 비교하여 참(True) 또는 거짓(False)을 반환합니다.

```python
x = 5
y = 10

print(x == y)  # 같음: False
print(x != y)  # 다름: True
print(x > y)   # 초과: False
print(x < y)   # 미만: True
print(x >= y)  # 이상: False
print(x <= y)  # 이하: True
```

#### 문자열 비교

```python
# 문자열 비교는 사전식 순서로 진행됨
print("apple" < "banana")  # True (a는 b보다 알파벳 순서상 앞)
print("abc" < "abd")       # True (동일한 위치의 글자 비교)
print("abc" < "abcd")      # True (길이가 짧은 문자열이 먼저)
```

#### 논리 연산자

여러 조건을 결합합니다.

```python
p = True
q = False

# and: 두 조건 모두 참일 때만 참
print(p and q)  # False

# or: 하나라도 참이면 참
print(p or q)   # True

# not: 참/거짓 반전
print(not p)    # False
print(not q)    # True
```

#### 복합 논리 조건

```python
age = 25
income = 35000

# 여러 조건 결합
is_eligible = (age > 18) and (income >= 30000)
print(is_eligible)  # True

# 우선순위: not > and > or
result = not age > 18 or income > 40000
print(result)  # False (not (age > 18) 가 False, income > 40000도 False)
```

#### 할당 연산자

변수에 값을 할당합니다.

```python
# 기본 할당
x = 10

# 복합 할당
x += 5   # x = x + 5와 동일, x는 15가 됨
print(x)  # 15

x -= 3   # x = x - 3와 동일, x는 12가 됨
print(x)  # 12

x *= 2   # x = x * 2와 동일, x는 24가 됨
print(x)  # 24

x /= 4   # x = x / 4와 동일, x는 6.0이 됨
print(x)  # 6.0

x //= 2  # x = x // 2와 동일, x는 3.0이 됨
print(x)  # 3.0

x %= 2   # x = x % 2와 동일, x는 1.0이 됨
print(x)  # 1.0

y = 2
y **= 3  # y = y ** 3와 동일, y는 8이 됨
print(y)  # 8
```

#### 멤버십 연산자

시퀀스 내 값의 존재 여부를 확인합니다.

```python
fruits = ["apple", "banana", "cherry"]

# in: 포함되어 있으면 True
print("apple" in fruits)     # True
print("orange" in fruits)    # False

# not in: 포함되어 있지 않으면 True
print("mango" not in fruits) # True
```

#### 식별 연산자

객체의 메모리 주소를 비교합니다.

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

# is: 동일한 객체를 참조하면 True
print(a is c)  # True (c는 a와 같은 객체를 참조)
print(a is b)  # False (내용은 같지만 다른 객체)

# is not: 다른 객체를 참조하면 True
print(a is not b)  # True
```

#### **주의사항**

#### 1. 정수 나눗셈: 나눗셈 결과를 정수로 얻으려면 // 연산자를 사용합니다. 일반 나눗셈(/)은 항상 실수를 반환합니다.

```python
print(10 / 5)   # 2.0 (float)
print(10 // 5)  # 2 (int)
```

#### 2. 부동소수점 연산 오차: 컴퓨터는 실수를 이진 부동소수점으로 표현하기 때문에 작은 오차가 발생할 수 있습니다.

```python
print(0.1 + 0.2)  # 0.30000000000000004 (정확히 0.3이 아님)
```

#### 3. 비교 연산자 체이닝: 파이썬에서는 여러 비교 연산자를 연결할 수 있습니다.

```python
x = 5
# 아래 두 표현은 동일함
print(1 < x < 10)       # True
print(1 < x and x < 10) # True
```

#### 4. 단락 평가(Short-circuit evaluation): and와 or 연산자는 왼쪽에서 오른쪽으로 평가하며 결과가 확정되면 나머지 부분은 평가하지 않습니다.

```python
# or 연산에서 첫 번째 조건이 True면 두 번째 조건은 평가하지 않음
result = True or print("이 문장은 실행되지 않습니다")

# and 연산에서 첫 번째 조건이 False면 두 번째 조건은 평가하지 않음
result = False and print("이 문장도 실행되지 않습니다")
```
